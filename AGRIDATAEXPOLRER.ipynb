```python
#Impact of Area Cultivated on Production (Rice, Wheat, Maize)
columns_of_interest = [
    'State Name', 
    'RICE AREA (1000 ha)', 'RICE PRODUCTION (1000 tons)', 
    'WHEAT AREA (1000 ha)','WHEAT PRODUCTION (1000 tons)', 
    'MAIZE AREA (1000 ha)', 'MAIZE PRODUCTION (1000 tons)'
]
crop_data = df[columns_of_interest]
plt.figure(figsize=(18, 6))
# Rice Production vs Area Cultivated
plt.subplot(1, 3, 1)
sns.scatterplot(x=crop_data['RICE AREA (1000 ha)'], y=crop_data['RICE PRODUCTION (1000 tons)'], color='blue')
plt.title('Rice Production vs Area Cultivated', fontsize=18,fontweight='bold')
plt.xlabel('Area Cultivated (1000 ha)', fontsize=14,fontweight='bold')
plt.ylabel('Rice Production (1000 tons)', fontsize=14,fontweight='bold')
# Wheat Production vs Area Cultivated
plt.subplot(1, 3, 2)
sns.scatterplot(x=crop_data[ 'WHEAT AREA (1000 ha)'], y=crop_data['WHEAT PRODUCTION (1000 tons)'], color='green')
plt.title('Wheat Production vs Area Cultivated', fontsize=18,fontweight='bold')
plt.xlabel('Area Cultivated (1000 ha)', fontsize=14,fontweight='bold')
plt.ylabel('Wheat Production(1000 tons)', fontsize=14,fontweight='bold')
# Maize Production vs Area Cultivated
plt.subplot(1, 3, 3)
sns.scatterplot(x=crop_data[ 'MAIZE AREA (1000 ha)'], y=crop_data['MAIZE PRODUCTION (1000 tons)'], color='orange')
plt.title('Maize Production vs Area Cultivated', fontsize=18,fontweight='bold')
plt.xlabel('Area Cultivated (1000 ha)', fontsize=14,fontweight='bold')
plt.ylabel('Maize Production (1000 tons)', fontsize=14,fontweight='bold')
plt.tight_layout()
plt.show()
# Step 4: Calculate correlation between Area and Production for each crop
rice_corr = crop_data['RICE AREA (1000 ha)'].corr(crop_data['RICE PRODUCTION (1000 tons)'])
wheat_corr = crop_data['WHEAT AREA (1000 ha)'].corr(crop_data['WHEAT PRODUCTION (1000 tons)'])
maize_corr = crop_data['MAIZE AREA (1000 ha)'].corr(crop_data['MAIZE PRODUCTION (1000 tons)'])

# Step 5: Print the correlation values
print(f'Correlation between Rice Area and Production: {rice_corr}')
print(f'Correlation between Wheat Area and Production: {wheat_corr}')
print(f'Correlation between Maize Area and Production: {maize_corr}')
```

```python
import pandas as pd
import pymysql

# Establish connection
myconnection = pymysql.connect(host='127.0.0.1', user='root', passwd='sukil')

# Load DataFrame
df = pd.read_csv("C://Users//SukilPriya//Downloads//ICRISAT-District Level Data - ICRISAT-District Level Data.csv")

# Get database and table name from user input
database_name = input("Enter Your Database Name: ")
table_name = input("Enter Your Table Name: ")

# Data type mapping for MySQL
dtype_mapping = {
    'int64': 'INTEGER',
    'float64': 'FLOAT',
    'object': 'TEXT',
    'datetime64[ns]': 'DATETIME',
    'bool': 'BOOLEAN'
}

# Create cursor to interact with MySQL
cursor = myconnection.cursor()

# Check if the database exists, if not, create it
cursor.execute(f"SHOW DATABASES LIKE '{database_name}';")
result = cursor.fetchone()
if result is None:
    print(f"Database {database_name} does not exist. Creating the database...")
    cursor.execute(f"CREATE DATABASE `{database_name}`;")
    myconnection.commit()

# Now, switch to the database
cursor.execute(f"USE `{database_name}`;")

# Construct the columns definition for CREATE TABLE query
columns = []
for col, dtype in zip(df.columns, df.dtypes):
    column_name = f"`{col}`"  # Wrap column name in backticks for special characters or spaces
    column_type = dtype_mapping.get(str(dtype), 'TEXT')  # Default to TEXT for unsupported types
    columns.append(f"{column_name} {column_type}")

columns_str = ",".join(columns)

# Construct the CREATE TABLE query
create_table_query = f"CREATE TABLE `{table_name}` ({columns_str});"

# Execute the CREATE TABLE query
try:
    cursor.execute(create_table_query)
except pymysql.MySQLError as e:
    print(f"Error creating table: {e}")
    myconnection.rollback()
    exit()

# Insert data into the table using parameterized queries
insert_query = f"INSERT INTO `{table_name}` ({','.join([f'`{col}`' for col in df.columns])}) VALUES ({','.join(['%s'] * len(df.columns))})"

# Insert each row in the DataFrame
for i in range(len(df)):
    cursor.execute(insert_query, tuple(df.iloc[i]))

# Commit the transaction
myconnection.commit()

print("Table created and data inserted successfully.")
```
